```python
names_AMV = ['ECT2N','ECT2N_del2','ECT2N_r2_9xYA','A1','FUS']
clean_names_AMV = ['ECT2',r'ECT2 $\Delta$5','ECT2 9x Y->A','A1','FUS']

names_all = []
for n in names_AMV:
    names_all.append(n)

temp = 293
fbase = '/Users/sobuelow/projects/ect2/slab/ect2_variants'
aminoacids = "ACGILMNPQSTVDEKRHFWY"
residues = pd.read_csv('residues.csv').set_index('one')
df_proteins = initProteins()
for name in names_AMV:
    df_proteins = addProtein(df_proteins,name,ffasta='fastabib.fasta')

plt.rcParams['font.size'] = 14
```

```python
Lxys = {}
for name in names_all:
#     print(name)
    path = f'{fbase}/{name}/{temp:d}'
    u = mda.Universe(f'{path}/top.pdb')
    print(name, u.dimensions / 10.)
    if 'ECT2N' in name:
        Lxys[name] = u.dimensions[0] / 10. # correct for inconsistent top.pdb
    else:
        Lxys[name] = u.dimensions[0]
```

# Center slabs

```python
temps = [293]

for name in names_all:
    for temp in temps:
        print(name, temp)
        path = f'{fbase}/{name}/{temp:d}'
        center_slab(path,name,temp,start=1,end=None,step=1)
```

```python
def calc_mw(fasta):
    x = "".join(fasta)
    mw = Bio.SeqUtils.molecular_weight(x,seq_type='protein')
    return mw
```

```python
plt.rcParams['font.size'] = 14

temp = 293
# name = 'ECT2N_r2_6xYA'

names = names_AMV
clean_names = clean_names_AMV
fig, ax = plt.subplots(1,5,figsize=(16,8),sharey=True)

for idx,name in enumerate(names):
    print('------------')
    clean_name = clean_names[idx]
    print(clean_name)
    axij = ax[idx]
    ax[0].set(ylabel='Timestep')
    ymax = 7500

    f = df_proteins.loc[name].fasta
    N = len(f)
    mw = calc_mw(f) # Da
    print(mw)
    cmap=plt.cm.Blues
    hs = np.load(f'{fbase}/{name}/{temp:d}/{name}_{temp:d}.npy')
    L = Lxys[name] / 10. # in nm
    conv = 100/6.022/N/L/L*1e3 # convert to protein concentration [mmol/L]
    hconv_mM = hs * conv
    hconv_mgml = hconv_mM * 1e-3 * mw # convert to protein concentration [mg/mL]
    print(np.max(hconv_mgml))
    print(hconv_mgml.shape)
#     print(np.min(hconv),np.max(hconv))
    print(hconv_mgml[0:ymax-10:10,::10].shape)
    axij.imshow(hconv_mgml[0:ymax-10:10,::10],cmap=cmap,vmin=0,vmax=700,
               aspect='auto')
    
    axij.set_title(clean_name,fontsize=14)#,rotation=90)
    
    axij.set(xlabel=r'z [nm]')
    
    if ('ECT' in name) or ('YTHDF' in name):# or ('MRB1' in name):
        xticks = [50,150,250]
        xticklabels = [-100,0,100]
    else:
        xticks = [25,75,125]
        xticklabels = [-50,0,50]
    axij.set_xticks(xticks)
    axij.set_xticklabels(xticklabels)
    axij.set_yticks(np.arange(0,ymax/10,100))
    axij.set_yticklabels(np.arange(0,ymax,1000))

fig.tight_layout(w_pad=1.0)
# fig.savefig(f'figures/slab_timetraces.pdf')
```

```python
temp = 293

tmin = 2000
tmax = 7500
names = names_AMV
clean_names = clean_names_AMV
colors = iter([plt.cm.tab10(i) for i in range(10)])

fig, ax = plt.subplots(figsize=(10,6))#,sharey=True)
    
for idx, name in enumerate(names):
    clean_name = clean_names[idx]
    axij = ax
    L = Lxys[name] / 10.
    f = df_proteins.loc[name].fasta
    mw = calc_mw(f) # Da
    print(mw)
    N = len(f)
    print(name)
    print(f'L={L}, n(AA)={N}')
    hs = np.load(f'{fbase}/{name}/{temp:d}/{name}_{temp:d}.npy')
    
    if hs.shape[1] == 3000:
        hs = hs[:,750:2250]

    conv = 100/6.022/N/L/L*1e3 # convert to protein concentration [mmol/L]
    hconv_mM = hs * conv
    hconv_mgml = hconv_mM * 1e-3 * mw # convert to protein concentration [mg/mL]
    
    print(hs.shape)
    print(f'total protein conc [mM]: {np.mean(hconv_mM):.2f}')
    print(f'total AA conc [mM]: {np.mean(hconv_mM)*N:.2f}')
    xs = np.arange(0,hs[0].shape[0]/10.,0.1)
    axij.plot(xs,np.mean(hconv_mgml[tmin:tmax],axis=0),
              label=clean_name,lw=1.0,c=next(colors))#colors[idx],lw=1.0)

axij = ax    
axij.set(xlim=(0,max(xs)+0.1))
axij.set(yscale='log')
axij.set(xlabel='z [nm]',ylabel='Density [mg/mL]')
axij.set_xticks(np.arange(15,136,20))
axij.set_xticklabels(np.arange(-60,61,20))
axij.legend(loc=1,fontsize=10,ncol=2)
fig.tight_layout()
# fig.savefig(f'figures/slab_concprofiles.pdf')
```

# Saturation concentrations

```python
tmin = 2000
tmax = 7500
names = names_AMV
clean_names = clean_names_AMV
colors = [plt.cm.tab10(i) for i in range(10)]
```

```python
temp = 293

vv = pd.DataFrame(index=names,dtype=object)
ee = vv.copy()

for name in names:
    L = Lxys[name] / 10.
    print(name,L)
    calcProfile(df_proteins,name,temp,L,vv,ee,fbase=fbase,tmin=tmin,tmax=tmax)
    print(len(df_proteins.loc[name].fasta))
print('values')
print(vv)
print('errors')
print(ee)

vv.to_pickle('conc_value.pkl')
ee.to_pickle('conc_error.pkl')
```

```python
T = 293.

fig, ax = plt.subplots(figsize=(10,6))#,sharey=True)

ymax = None

for idx,name in enumerate(names):
    print(name)
    clean_name = clean_names[idx]
    N = len(df_proteins.loc[name].fasta)
    
    c_dil = vv.loc[name][f'{temp}_dil']
    c_den = vv.loc[name][f'{temp}_den']
    
    e_dil = ee.loc[name][f'{temp}_dil']
    e_den = ee.loc[name][f'{temp}_den']
    
    dG_transfer = np.log(c_dil/c_den)
    
    ndraws = 10000
    
    spread_dil = np.random.normal(c_dil,e_dil,size=ndraws)
    spread_den = np.random.normal(c_den,e_den,size=ndraws)
    
    spread_dGs = np.zeros(ndraws)
    for idraw, (dil,den) in enumerate(zip(spread_dil,spread_den)):
        spread_dGs[idraw] = np.log(dil/den)
        
    dG_error = np.std(spread_dGs)
    print(dG_transfer,dG_error)
    
    if name == 'ECT2N':
        ax.axhline(dG_transfer,c='gray',ls='dotted')
    ax.errorbar(idx,dG_transfer,yerr=dG_error,marker='o',c='black',capsize=5)

ax.set(ylabel='$\Delta G_T$ [k$_B$T]')

ax.set_xticks(np.arange(len(vv[f'{temp}_dil'])))
ax.grid(ls='dotted',alpha=0.4)
ax.set_xticklabels(clean_names,rotation=90)
fig.tight_layout()
# fig.savefig(f'figures/dG.pdf')
```
